```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
kc_house_data=pd.read_csv("/content/drive/MyDrive/NLP dataset/kc_house_data.csv")
kc_house_data.head()
```

```python
conditions = [
    (kc_house_data['yr_built'] >= 1970) & (kc_house_data['yr_built'] <= 2000),
    (kc_house_data['yr_built'] >= 2001) & (kc_house_data['yr_built'] <= 2010),
    (kc_house_data['yr_built'] >= 2011) & (kc_house_data['yr_built'] <= 2015),
]

categories = ["Old Construction", "Moderate Construction", "New Construction"]

kc_house_data['construction_category'] = np.select(conditions, categories, default="Other")
print(kc_house_data)
```

```python
filtered_data = kc_house_data[
    (kc_house_data['yr_built'] >= 2003) &
    (kc_house_data['yr_built'] <= 2012) &
    (kc_house_data['bedrooms'] == 2) &
    (kc_house_data['floors'] == 1)
]

print(filtered_data)

```

```python

filtered_data = kc_house_data[
    (kc_house_data['yr_built'] >= 2003) &
    (kc_house_data['yr_built'] <= 2012) &
    (kc_house_data['bedrooms'] == 2) &
    (kc_house_data['floors'] == 1)
]

house_prices = filtered_data['price']

plt.figure(figsize=(10, 6))
sns.histplot(house_prices, bins=30, kde=True, color='blue')
plt.title('Histogram Distribution of House Prices', fontsize=16)
plt.xlabel('House Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(8, 8))
stats.probplot(house_prices, dist="norm", plot=plt)
plt.title('QQ Plot of House Prices', fontsize=16)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python

price_relation = kc_house_data.groupby(['construction_category', 'floors'])['price'].mean().reset_index()

print(price_relation)


plt.figure(figsize=(12, 6))
sns.boxplot(data=kc_house_data, x='construction_category', y='price', hue='floors', palette='Set2')
plt.title('Price Distribution by Construction Category and Floors', fontsize=16)
plt.xlabel('Construction Category', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.legend(title='Floors', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


plt.figure(figsize=(12, 6))
sns.scatterplot(data=kc_house_data, x='floors', y='price', hue='construction_category', palette='Set1')
plt.title('Price vs Floors by Construction Category', fontsize=16)
plt.xlabel('Floors', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.legend(title='Construction Category')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python


columns_to_analyze = ['price', 'sqft_above', 'sqft_lot', 'sqft_living']

outlier_info = {}

for column in columns_to_analyze:

    Q1 = kc_house_data[column].quantile(0.25)
    Q3 = kc_house_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = kc_house_data[(kc_house_data[column] < lower_bound) | (kc_house_data[column] > upper_bound)]

    outlier_info[column] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outliers Count': len(outliers)
    }

    plt.figure(figsize=(10, 6))
    sns.boxplot(x=kc_house_data[column], color='skyblue')
    plt.title(f'Boxplot of {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

for column, stats in outlier_info.items():
    print(f"\nColumn: {column}")
    for key, value in stats.items():
        print(f"{key}: {value}")
```
